```python
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
```

```python
#define the features
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
forest = RandomForestClassifier(n_estimators=1000, n_jobs=-1, class_weight='balanced')
```

```python
# define Boruta feature selection method
feat_selector = BorutaPy(forest, n_estimators='auto', verbose=2)
```

```python
# find all relevant features
feat_selector.fit(X, y)
```

```python
# show the most important features
most_important = data.columns[:-1][feat_selector.support_].tolist()
most_important
```

```python
# select the top 6 features
top_features = data.columns[:-1][feat_selector.ranking_ <6].tolist()
top_features
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from collections import Counter
```

```python
X = data[top_features]
y = data.iloc[:,-1]
```

```python
# the numbers before smote
num_before = dict(Counter(y))

#perform smoting
# Specify sampling strategies as a dictionary for multi-class
sampling_strategy = {
    0: 160,  # Replace with the desired number of samples for class 0
    1: 54,  # Replace with the desired number of samples for class 1
    2: 35,# Replace with the desired number of samples for class 2, etc.
    3: 35,
    4: 13,
}

# Define pipeline with SMOTE and RandomUnderSampler
over = SMOTE(sampling_strategy=sampling_strategy, random_state=42)
under = RandomUnderSampler(sampling_strategy=sampling_strategy, random_state=42)
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)

# Transform the dataset
X_smote, y_smote = pipeline.fit_resample(X, y)

# Check the class distribution after SMOTE
num_after = dict(Counter(y_smote))
print(f'Class distribution after SMOTE: {num_after}')
```

```python
print(num_before, num_after)
```

```python
new_data = pd.concat([pd.DataFrame(X_smote), pd.DataFrame(y_smote)], axis=1)
new_data.columns = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'ca', 'num']  # Ensure all columns are included
new_data.head()
```

```python
x1 = new_data[top_features]
y1= new_data.iloc[:,-1]
x1.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x1, y1,test_size=.2,
                                                 random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
# Calculate accuracy
nb_accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {nb_accuracy:.2f}')
```

```python
# Print classification report
print(classification_report(y_test, y_pred))
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y1, test_size=0.2,
                                                    random_state=42)
```

```python
# Define the model
model = Sequential()
model.add(Dense(units=64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))  # For binary classification
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
```

```python
# Evaluate the model
y_pred = model.predict(X_test)
y_pred_binary = (y_pred > 0.5).astype(int)  # Threshold for binary classification
```

```python
# Accuracy
ann_accuracy = accuracy_score(y_test, y_pred_binary)
print('Accuracy:', ann_accuracy)
```
