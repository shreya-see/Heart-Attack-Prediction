## HEART ATTACK RISK PREDICTION

Risk of Heart Attack is increasing day by day. This model here helps to predict the risk of heart attack and our model here tells the level of heart attack the person is suffering from. Here we have levels from 0 to 4, more the level, more will be the risk of heart attack in a person.

The person is supposed to mention a few features like age, cholestrol etc and based on these features our model will predict the risk of heart attack a person has.

### 4. Feature Selection

```python
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
```

```python
#define the features
X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
forest = RandomForestClassifier(n_estimators=1000, n_jobs=-1, class_weight='balanced')
```

```python
# define Boruta feature selection method
feat_selector = BorutaPy(forest, n_estimators='auto', verbose=2)
```

```python
# find all relevant features
feat_selector.fit(X, y)
```

```python
# show the most important features
most_important = data.columns[:-1][feat_selector.support_].tolist()
most_important
```

```python
# select the top 6 features
top_features = data.columns[:-1][feat_selector.ranking_ <=6].tolist()
top_features
```

```python
import statsmodels.api as sm
```

```python
X_top = data[top_features]
y = (data['num'] > 0).astype(int)
```

```python
res = sm.Logit(y, X_top).fit()
print(res.summary())
```

```python
params = res.params
conf = res.conf_int()
conf['Odds Ratio'] = params
conf.columns = ['5%', '95%', 'Odds Ratio']
print(np.exp(conf))
```

### 5. SMOTE

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from collections import Counter
```

```python
X = data[top_features]
y = data.iloc[:,-1]
```

```python
# the numbers before smote
num_before = dict(Counter(y))

#perform smoting
# Specify sampling strategies as a dictionary for multi-class
sampling_strategy = {
    0: 500,  # Replace with the desired number of samples for class 0
    1: 500,  # Replace with the desired number of samples for class 1
    2: 500,  # Replace with the desired number of samples for class 2, etc.
    # Add other classes if needed
}

# Define pipeline with SMOTE and RandomUnderSampler
over = SMOTE(sampling_strategy=sampling_strategy, random_state=42)
under = RandomUnderSampler(sampling_strategy=sampling_strategy, random_state=42)
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)

# Transform the dataset
X_smote, y_smote = pipeline.fit_resample(X, y)

# Check the class distribution after SMOTE
num_after = dict(Counter(y_smote))
print(f'Class distribution after SMOTE: {num_after}')
```

```python
print(num_before, num_after)
```

#### 5.1 Splitting data into Training and Testing set

```python
# new dataset
new_data = pd.concat([pd.DataFrame(X_smote), pd.DataFrame(y_smote)], axis=1)
new_data.columns = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'ca', 'thal','num']
new_data.head()
```

```python
x1 = new_data[top_features]
y1= new_data.iloc[:,-1]
x1.head()
```

```python
# x1 = data[['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'ca', 'thal']]
x1 = data[top_features]
y1 = data['num']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x1, y1,test_size=.2,
                                                 random_state=42)
```

### 6. Prediction Model

#### 6.1 Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import mean_squared_error
# Calculate accuracy
nb_accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {nb_accuracy:.2f}')
```

```python
from sklearn.metrics import mean_squared_error, matthews_corrcoef, classification_report
```

```python
# Calculate Mean Squared Error (MSE)
nb_mse = mean_squared_error(y_test, y_pred)
print(f'MSE: {nb_mse:.2f}')

# Calculate Root Mean Squared Error (RMSE)
nb_rmse = np.sqrt(nb_mse)
print(f'RMSE: {nb_rmse:.2f}')

# Calculate Matthews Correlation Coefficient (MCC)
nb_mcc = matthews_corrcoef(y_test, y_pred)
print(f'Matthews Correlation Coefficient (MCC): {nb_mcc:.2f}')

# Additionally, you can print the classification report
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

#### 6.3 Artificial Neural Network

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y1, test_size=0.2,
                                                    random_state=42)
```

```python
# Define the model
model = Sequential()
model.add(Dense(units=64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))  # For binary classification
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
```

```python
# Evaluate the model
y_pred = model.predict(X_test)
y_pred_binary = (y_pred > 0.5).astype(int)  # Threshold for binary classification
```

```python
# Accuracy
ann_accuracy = accuracy_score(y_test, y_pred_binary)
print('Accuracy:', ann_accuracy)
```

```python
# Calculate Mean Squared Error (MSE)
ann_mse = mean_squared_error(y_test, y_pred)
print(f'MSE: {ann_mse:.2f}')

# Calculate Root Mean Squared Error (RMSE)
ann_rmse = np.sqrt(ann_mse)
print(f'RMSE: {ann_rmse:.2f}')

# Ensure y_test and y_pred are categorical (e.g., integers or labels for classes)
if y_test.dtype in ['float64', 'float32']:
    y_test = y_test.astype(int)
if y_pred.dtype in ['float64', 'float32']:
    y_pred = y_pred.astype(int)

# Calculate Matthews Correlation Coefficient (MCC)
ann_mcc = matthews_corrcoef(y_test, y_pred)
print(f'Matthews Correlation Coefficient (MCC): {ann_mcc:.2f}')

# Additionally, you can print the classification report
print('Classification Report:')
print(classification_report(y_test, y_pred))
```
