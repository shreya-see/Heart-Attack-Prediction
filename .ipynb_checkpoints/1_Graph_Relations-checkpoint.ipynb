## 1. Exploratory Data Analysis

Exploratory Data Analysis (EDA) is a crucial step in data science where analysts dive deep into a dataset to understand its characteristics, uncover patterns, and identify relationships between variables.

It's like exploring a new territory without a map, relying on your senses and observations to make discoveries. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("framingham.csv")
```

```python
data = data.dropna()
```

```python
data.shape
```

### Histogram

```python
data_without_chd = data.drop(columns=['TenYearCHD'])
data_without_chd.hist(bins=10, figsize=(18, 20))
plt.show()
```

```python
data['TenYearCHD'].hist(color='orange', edgecolor='blue', figsize=(6, 4), rwidth=0.9)
plt.xlabel('TenYearCHD')
plt.ylabel('Frequency')
plt.title('Histogram to show Frequency vs TenYearCHD')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame and 'category_column' is your categorical variable
category_counts = data['category_column'].value_counts(normalize=True)  # Get proportions

# Bar Plot
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.xlabel('Categories')
plt.ylabel('Proportion')
plt.title('Proportion of Categories')
plt.show()
```
